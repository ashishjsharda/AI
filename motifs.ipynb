```python
import numpy as np
from tslearn.clustering import TimeSeriesKMeans
from tslearn.datasets import CachedDatasets
from tslearn.preprocessing import TimeSeriesScalerMeanVariance

# Load trace dataset
X_train, y_train, X_test, y_test = CachedDatasets().load_dataset("Trace")

# Normalize the time series data
X_train = TimeSeriesScalerMeanVariance().fit_transform(X_train)
X_test = TimeSeriesScalerMeanVariance().fit_transform(X_test)


n_clusters = 3  # Number of motifs to find
model = TimeSeriesKMeans(n_clusters=n_clusters, verbose=True, random_state=0)
y_pred = model.fit_predict(X_train)

# Extract and display motifs
for cluster_idx in range(n_clusters):
    cluster_samples = X_train[y_pred == cluster_idx]
    print(f"Motif for Cluster {cluster_idx}:")
    print(cluster_samples[0])  # Display the first sample in the cluster as a motif
```
